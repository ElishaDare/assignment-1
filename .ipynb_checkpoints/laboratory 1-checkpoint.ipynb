```python
import cv2
#read the image
image_path='imade for dip.jpeg'
image = cv2.imread('imade for dip.jpeg')
image
```

```python
import matplotlib.pyplot as plt 
import matplotlib.image as img 

plt.imshow(image)

plt.imshow(image) # display the image
plt.axis('off')
plt.show()
```

```python
#THIS IS THE IMAGE BUT IN BLUE SCALE 
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import os 
output_folder ='output_images'
outputpath = os.path.join(output_folder, 'img.jpeg')
cv2.imwrite(outputpath,image)
```

```python
# THIS IS A DIGITAL IMAGE BECAUSE IT CONTAINS PIXELS. 
#EACH PIXEL REPRESENTS A SINGLE POINT IN THE IMAGE AND HAS A NUMERICAL VALUE REPRESENTING COLOUR OR INTENSITY.
#WE USED OPENCV TO DISPLAY THE IMAGE THIS WITHOUT A DOUBT MAKES IT A DIGITAL IMAGE.
#AS LONG AS AN IMAGE IS STORED AND PROCESSED IN A DIGITAL FORMAT IT IS A DIGITAL IMAGE.
```

```python
# QUESTION 2
#GREY SCALE 
```

```python
gsimage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(gsimage, cv2.COLOR_BGR2RGB)) # display the image
plt.axis('off')
plt.show()
```

```python
output_folder ='output_images'
outputpath = os.path.join(output_folder, 'Grey_image.jpeg')
cv2.imwrite(outputpath,gsimage)
```

```python
# AN IMAGE IN GREYSCALE IS STILL CONSIDERED A DIGITAL IMAGE. 
#THIS IS BECAUSE REGARDLESS THE IMAGE HAS BEEN CONVERTED TO DIGITAL FORMAT. REGARDLESS OF IT BEING GREY OR COLORED.
#IT TYPICALLY CONSISTS OF PIXELS AND EACH PIXEL REPRESENTS A SINGLE POINT IN THE IMAGE. 
#PIXEL VALUES IN A GREYSCALE REPRESENTS DIFFERENT POINTS SHOWING DIFFERENT INTENSITIES RANGING FROM BLACK TO WHITE.
#WHEN CONVERTED TO GREYSCALE COLOR IS REMOVED AND ALL THAT IS LEFT IS BLACK AND WHITE BUT AT DIFFERENT INTENSITIES.
#EACH PIXEL IN THE IMAGE STILL HAS NUMERICAL VALUES REPRESENTING INTENSITY MAKING IT A DIGITAL IMAGE.
```
